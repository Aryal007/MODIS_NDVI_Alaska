```python
from pathlib import Path
from utils import reproject_tiff, vis_tiff
import glob, os, rasterio
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
tiffs = ["anwr", "brw_atq_ivotuk", "prudhoebay"]
shps = ["ANWR.shp", "Brw_Atq_Ivotuk.shp", "PrudhoeBay.shp"]
selected = 1

DATA_DIR = Path(f"/datadrive/MODIS_NDVI_Alaska/modis/tiffs/{tiffs[selected]}")
SHAPEFILE_DIR = Path(f"/datadrive/MODIS_NDVI_Alaska/modis/shps/{shps[selected]}")
files = glob.glob(str(DATA_DIR / "*"))
files.sort()
label = gpd.read_file(SHAPEFILE_DIR)
dst_crs = rasterio.crs.CRS.from_string(str(label.crs))
```

```python
OUT_DIR = Path(f"/datadrive/MODIS_NDVI_Alaska/modis/tiffs_reproject/{tiffs[selected]}")
if not os.path.exists(OUT_DIR):
    os.makedirs(OUT_DIR, exist_ok=True)
```

### Before reprojection

```python
print(f"CRS: {rasterio.open(files[1]).crs}")
vis_tiff(files[50])
```

```python
for i, f in enumerate(files):
    print(f"File {i+1} out of {len(files)}")
    reproject_tiff(f, dst_crs, OUT_DIR)
```

```python
files = glob.glob(str(OUT_DIR / "*"))
files.sort()
```

### After reprojection

```python
print(f"CRS: {rasterio.open(files[1]).crs}")
vis_tiff(files[50])
```
